```python
import pandas as pd
```

```python
df=pd.read_csv("../data/temperature.csv")
df
```

```python
df=df.drop("record_id", axis=1)
df
```

```python
uniqueCities=df["City"].unique()
print(uniqueCities)
uniqueCountries=df["Country"].unique()
print(uniqueCountries)
uniqueAverageTemperatureFahr=df["AverageTemperatureFahr"].unique()
print(uniqueAverageTemperatureFahr)
uniqueAverageTemperatureUncertaintyFahrs=df["AverageTemperatureUncertaintyFahr"].unique()
print(uniqueAverageTemperatureUncertaintyFahrs)
```

```python
df=df.dropna(subset=["City","Country","AverageTemperatureFahr","AverageTemperatureUncertaintyFahr"])
df
```

```python
df=df.drop("day", axis=1)
df
```

```python
df["AverageTemperatureCel"]=(df["AverageTemperatureFahr"]-32)*5/9
df["AverageTemperatureUncertaintyCel"]=(df["AverageTemperatureUncertaintyFahr"]-32)*5/9
df=df.drop(["AverageTemperatureUncertaintyFahr","AverageTemperatureFahr"],axis=1)
df
```

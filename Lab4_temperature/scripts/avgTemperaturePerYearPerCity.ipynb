```python
import pandas as pd
```

```python
#loading data
df=pd.read_csv("../data/cleaned_temperature.csv")
df
```

```python
#cleaning data
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

```python
#for each country creating a csv file with pairs in columns: year, avg temp in this year

for city in df["City"].unique():
        
    dfCity=df[df["City"]==city]

    countryCode=dfCity.iloc[1]['country_id']
    file_name="../data/cities/"+str(city)+"_"+str(countryCode)+"_avg_temp.csv"

    dictionary= dict()
    for year in dfCity["year"].unique():
        dfCityYear=dfCity[dfCity["year"]==year]
        avgTemp=dfCityYear["AverageTemperatureCel"].mean()
        l=[avgTemp]
        dictionary[year]=l
            
    dfToSave=pd.DataFrame(dictionary)
    dfToSave=dfToSave.sort_index(axis=1)
    dfToSave.to_csv(file_name,index=False)
```

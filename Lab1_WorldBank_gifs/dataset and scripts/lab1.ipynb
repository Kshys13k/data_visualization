```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_85.csv')
df.head()
```

```python
sortedDf= df.sort_values(by="2022", ascending=False)
sortedDf.head(50)
```

```python
dfIndia=df.iloc[109,-63:-1]
dfChina=df.iloc[40,-63:-1]
dfIndonesia=df.iloc[106,-63:-1]
dfPakistan=df.iloc[184,-63:-1]
dfNigeria=df.iloc[174,-63:-1]
dfForPlot=pd.concat([dfIndia, dfChina, dfIndonesia, dfPakistan, dfNigeria], axis=1)
dfForPlot.columns=["India","China","Indonesia","Pakistan","Nigeria"]

dfForPlot.head()
```

```python
dfForPlot.tail()
```

```python
#One plot
row = dfForPlot.iloc[0]

plt.figure(figsize=(10,6))
row.plot(kind="bar")

plt.xticks(ticks=range(len(row.index)), labels=row.index)

plt.title('Comparison of the populations of the largest countries')
plt.xlabel('Country')
plt.ylabel('Population')

plt.show()

#Second plot
row = dfForPlot.iloc[50]

plt.figure(figsize=(10,6))
row.plot(kind="bar")

plt.xticks(ticks=range(len(row.index)), labels=row.index)

plt.title('Comparison of the populations of the largest countries')
plt.xlabel('Country')
plt.ylabel('Population')

plt.show()
```

```python
#Generate all plots for gif

for index, row in dfForPlot.iterrows():
    plt.figure(figsize=(10,6))
    row.plot(kind="bar")
    
    plt.xticks(ticks=range(len(row.index)), labels=row.index)

    title= "Comparison of the populations of the largest countries " + str(index)
    plt.title(title)
    plt.xlabel('Country')
    plt.ylabel('Population')
    
    filename="plots/plot"+str(index)+".png"
    plt.savefig(filename)
    plt.close()
```

```python
#Function for generating plots
def plotGenerator(dfForPlot):
    for index, row in dfForPlot.iterrows():
        plt.figure(figsize=(10,6))
        row.plot(kind="bar")

        plt.xticks(ticks=range(len(row.index)), labels=row.index)

        title= "Comparison of the populations of the largest countries " + str(index)
        plt.title(title)
        plt.xlabel('Country')
        plt.ylabel('Population')

        filename="plots/plot"+str(index)+".png"
        plt.savefig(filename)
        plt.close()
```

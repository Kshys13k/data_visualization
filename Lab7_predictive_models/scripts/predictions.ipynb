```python
import pandas as pd
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error
```

# Brasil AR

```python
#selecting data 
df=pd.read_csv("../data/BRA_avg_temp.csv")
df
```

```python
#predictive model
LAGS=15
PREDICTION_YEARS=250
data=df[0:1].to_numpy().reshape(-1,1)
ar_model = AutoReg(data, LAGS)
fitted_model=ar_model.fit()
prediction=fitted_model.predict(0,len(data)+PREDICTION_YEARS)
```

```python
#plot
plt.figure(figsize=(5,5))
to_int=np.vectorize(int)
x=to_int(df.columns.to_numpy())
plt.scatter(x,data, s=3, label="Real data")
x1=np.append(x,(range(x[-1]+1,x[-1]+PREDICTION_YEARS+2)))
plt.scatter(x1,prediction, s=3, c="orange", label="Prediction")
plt.xlabel("year")
plt.ylabel("T [Cel]")
plt.legend(loc="lower right")

file_path=f"../plots/predictions/brasil_ar_{PREDICTION_YEARS}years.png"
plt.savefig(file_path)
```

```python
#MAE
tscv = TimeSeriesSplit(n_splits=3)
mae_values=[]
for train_index, test_index in tscv.split(data):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = data[train_index], data[test_index]
    X_train=X_train.reshape(-1,1)
    ar_model_test=AutoReg(X_train,LAGS)
    fitted_model_test=ar_model_test.fit()
    prediction_test=fitted_model_test.predict(0,len(data))
    
    #X_test - true values
    X_pred=prediction_test[test_index]
    mae=mean_absolute_error(X_test,X_pred)
    mae_values.append(mae)
    print(f"!!!!!!!!___MAE={mae}___!!!!!!!!")
    
average_mae=np.mean(mae_values)
print("AVERAGE MAE: ",average_mae)
```

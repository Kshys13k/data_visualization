# Changes in population during war in Rwanda

```python
#gathering data
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../data/API_SP.POP.TOTL_DS2_en_csv_v2_85.csv")
df.head()
```

```python
df=df.sort_values(by="2022", ascending=False)
df.head()
```

```python
Rwanda= df[df["Country Name"]=="Rwanda"]
Rwanda
```

```python
Burundi=df[df["Country Name"]=="Burundi"]
Burundi
```

```python
Uganda=df[df["Country Name"]=="Uganda"]
Uganda
```

```python
df=pd.concat([Rwanda,Burundi,Uganda],axis=0)
df
```

```python
df1=df.iloc[:,0]
df2=df.iloc[:,-63:]
df=pd.concat([df1,df2],axis=1)

df
```

```python
df.iloc[:,1:]=df.iloc[:,1:]/100000
df
```

```python
df.to_csv("../data/rwanda.csv", index=False, sep=",", encoding="utf-8")
```

```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

# Task 1

```python
df=pd.read_csv("../data/ans.csv")
```

```python
df
```

```python
plt.figure(figsize=(6,6))


for i in range (1,5):
    x=f"x{i}"
    y=f"y{i}"
    #values for table:
    print(f"___x{i}_y{i}___")
    print("Mean of X: ",df[x].mean())
    print("Mean of Y: ",df[y].mean())
    print("Sample variance of X: ",df[x].var())
    print("Sample variance of Y: ",df[y].var())
    print("Correlation between X and Y: ",np.corrcoef(df[x],df[y])[0,1])
    
    X=df[x].to_numpy().reshape(-1,1)
    Y=df[y].to_numpy()
    regr = linear_model.LinearRegression()
    regr.fit(X,Y)
    intercept=regr.intercept_
    coef= regr.coef_[0]
    print(f"Linear regression: y={intercept}+{coef}x")
    
    y_pred= regr.predict(X)
    print("R^2: ", r2_score(Y, y_pred))
    
    index=220+i
    plt.subplot(index)
    plt.plot(df[x],y_pred)
    plt.scatter(df[x],Y, marker="o", color="orange", edgecolors="red", linewidths=1)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.xlim(3,20)
    plt.ylim(2,14)

plt.subplots_adjust(wspace=0.3) 
plt.savefig("../plots/task1.png")
```

# Task 2

```python
df=pd.read_csv("../data/ans2.tsv", sep="\t")
df
```

```python
list_of_dfs=[]
for un in df["dataset"].unique():
    df_=df[df["dataset"]==un]
    list_of_dfs.append(df_)

list_of_dfs[0]
```

```python
plt.figure(figsize=(12,12))
values={"meansX": [], "meansY": [], "sdX": [], "sdY": [], "corr": []}

for i, df in enumerate(list_of_dfs):
    df=df.reset_index()
    dataset_name=df.loc[0,"dataset"]
    
    #values for table:
    print(f"___{dataset_name}___")
    print("Mean of X: ",df["x"].mean())
    print("Mean of Y: ",df["y"].mean())
    print("SD of X: ",df["x"].std())
    print("SD of Y: ",df["y"].std())
    print("Correlation between X and Y: ",np.corrcoef(df["x"],df["y"])[0,1])
    
    values["meansX"].append(df["x"].mean())
    values["meansY"].append(df["y"].mean())
    values["sdX"].append(df["x"].std())
    values["sdY"].append(df["y"].std())
    values["corr"].append(np.corrcoef(df["x"],df["y"])[0,1])
    
    plt.subplot(4,4,i+1)
    plt.scatter(df["x"],df["y"])
    plt.xlabel(x)
    plt.ylabel(y)

plt.subplots_adjust(wspace=0.3) 
plt.savefig("../plots/task2.png")

min_=min(values["meansX"])
max_=max(values["meansX"])
print(f"Means of x: ({min_},{max_}), diff/2: {(min_-max_)/2}, val: {(min_+max_)/2}")
min_=min(values["meansY"])
max_=max(values["meansY"])
print(f"Means of y: ({min_},{max_}),diff/2: {(min_-max_)/2}, val: {(min_+max_)/2}")
min_=min(values["sdX"])
max_=max(values["sdX"])
print(f"Vars of x: ({min_},{max_}), diff/2: {(min_-max_)/2}, val: {(min_+max_)/2}")
min_=min(values["sdY"])
max_=max(values["sdY"])
print(f"Vars of y: ({min_},{max_}), diff/2: {(min_-max_)/2}, val: {(min_+max_)/2}")
min_=min(values["corr"])
max_=max(values["corr"])
print(f"Correlation: ({min_},{max_}), diff/2: {(min_-max_)/2}, val: {(min_+max_)/2}")
```

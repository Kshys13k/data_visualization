```python
import pandas as pd
```

```python
#loading data
df=pd.read_csv("../data/cleaned_temperature.csv")
df
```

```python
#cleaning data
df.drop("Unnamed: 0", axis=1, inplace=True)
df
```

```python
#for each country creating a csv file with pairs in columns: year, avg temp in this year
for country in df["country_id"].unique():
    
    dfCountry=df[df["country_id"]==country] 
    
    file_name="../data/"+str(country)+"_avg_temp.csv"
    
    dictionary= dict()
    for year in dfCountry["year"].unique():
        dfCountryYear=dfCountry[dfCountry["year"]==year]
        avgTemp=dfCountryYear["AverageTemperatureCel"].mean()
        l=[avgTemp]
        dictionary[year]=l
    
    dfToSave=pd.DataFrame(dictionary)
    dfToSave=dfToSave.sort_index(axis=1)
    dfToSave.to_csv(file_name,index=False)
```
